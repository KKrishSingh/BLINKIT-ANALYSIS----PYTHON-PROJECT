## **Business Requirements**

#### **KPI's Requirements**

```python
#Total Sales

total_sales = df["Sales"].sum()

#Average Sales
average_sales = df["Sales"].mean()

#No of Items Sold
no_of_items_sold = df["Sales"].count()

#Average Ratings
average_rating = df["Rating"].mean()

#Display
print(f"Total sales : ${total_sales:,.0f}")
print(f"average_sales : ${average_sales:,.0f}")
print(f"no_of_items_sold : ${no_of_items_sold:,.0f}")
print(f"average_rating: {average_rating:,.1f}")
```

## **Charts Requirements**

#### **Total Sales By Fat Content**

```python
sales_by_fat = df.groupby('Item Fat Content')["Sales"].sum()

c = colors=["#808F85","#BFB48F"]
plt.pie(sales_by_fat , labels=sales_by_fat.index,autopct = "%.0f%%",startangle = 90 ,colors= c)
plt.title("Sales By Fat Content",fontsize=20)
plt.axis("equal")
plt.show()
```

ANALYSIS : The pie chart highlights that Low Fat items contribute the majority of total sales (65%), while Regular items account for only 35%. This indicates a clear customer preference for Low Fat products, making them the dominant category in overall sales. Businesses could leverage this trend by prioritizing Low Fat product offerings, promotions, and inventory management to maximize revenue.

#### **Total Sales by Item Type**

```python
## sales_by_type
sales_by_type = df.groupby("Item Type")["Sales"].sum().sort_values(ascending = False)

plt.figure(figsize=(10,6))
bars = plt.bar (sales_by_type.index,sales_by_type.values,color="#83BCA9")

plt.xticks (rotation = 90 )
plt.xlabel ("Item Type",fontsize=15)
plt.ylabel ("Sales",fontsize=15)
plt.title ("Total Sales By Item Type",fontsize=20)

for bar in bars:
    plt.text( bar.get_x() + bar.get_width() /2, bar.get_height(),
             f'{bar.get_height():,.0f}', ha = 'center' ,va ='bottom',fontsize=8)

plt.grid(True , alpha=0.3)
plt.tight_layout()
plt.show()
```

ANALYSIS : The chart shows total sales trends by outlet establishment year. Sales were highest in 1998 (204k) but dropped sharply by 1999 (131k) and hit the lowest point in 2009 (78k). From 2013 onwards, sales recovered and stabilized, staying consistently between 129k–133k. This suggests that while older outlets initially performed strongly, newer ones have maintained more stable but moderate sales levels over time.

#### **Fat Content By Outlet For Total Sales**

```python
grouped = df.groupby(["Outlet Location Type","Item Fat Content"])["Sales"].sum().unstack()
groupsd = grouped [["Regular","Low Fat"]]

ax = grouped.plot(kind='bar',figsize=(8,5))
plt.title ("Outlet Tier By Item Fat Content",fontsize=20)
plt.xlabel ("Outlet Location Tier",fontsize=15)
plt.ylabel ("Total Sales",fontsize=15)
plt.legend(title='Item Fat Content')

plt.grid(True , alpha=0.3) 
plt.tight_layout()
plt.show()
```

ANALYSIS : The Bar chart shows that Low Fat items consistently achieve higher total sales than Regular items across all outlet tiers. Sales increase progressively from Tier 1 to Tier 3, with Tier 3 outlets recording the highest overall sales, particularly for Low Fat products. This trend highlights a strong customer preference for Low Fat items and suggests that focusing on their availability and promotion, especially in higher-tier outlets, could further enhance overall sales performance.

#### **Total Sales By Outlet Establishment**

```python
sales_by_year = df.groupby ("Outlet Establishment Year")["Sales"].sum().sort_index()

plt.figure(figsize=(9,5))
plt.plot(sales_by_year.index,sales_by_year.values, marker='o' , linestyle='-')

plt.title("Outlet Establishment")
plt.xlabel("Outlet Establishment Year")
plt.ylabel("Total Sales")

for x,y in zip (sales_by_year.index, sales_by_year.values):
    plt.text(x,y,f'{y:,.0f}', ha='center',va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```

ANALYSIS : The line chart illustrates total sales trends by outlet establishment year. Sales peaked strongly in 1998 (204,522) but dropped significantly in the following years, reaching a low around 2009 (78,432). From 2013 onwards, sales stabilized around 130k–133k, showing steady but flat performance. This indicates that older outlets initially drove higher sales, while newer establishments have maintained moderate and consistent levels.

#### **Sales By Outlet Size**

```python
sales_by_size = df.groupby ('Outlet Size')['Sales'].sum()

plt.figure (figsize=(4,4))
c = color=["#BEA2C2","#3BA99C","#33658A"]
plt.pie(sales_by_size,labels=sales_by_size.index, autopct = '%1.1f%%', startangle = 90,colors=c)
plt.title("Outlet Title")
plt.tight_layout()
plt.show()
```

ANALYSIS:
The pie chart shows the sales distribution across different outlet sizes. Small outlets contribute the largest share at 37%, followed closely by Medium outlets at 42.3%, while High outlets account for only 20.7%. This indicates that medium-sized outlets are driving the majority of sales, while high-sized outlets underperform comparatively, which could guide future resource allocation and strategy.

#### **Sales By Outlet Location**

```python
sales_by_location = df.groupby ('Outlet Location Type')["Sales"].sum().reset_index()
sales_by_location = sales_by_location.sort_values('Sales',ascending = True)

plt.figure(figsize=(8,3))
ax = sns.barplot (x='Sales',y='Outlet Location Type',data=sales_by_location)

plt.title("Total Sales By Outlet Location Type",fontsize=18)
plt.xlabel("Total Sales",fontsize=13)
plt.ylabel ("Outlet Location",fontsize=13)

plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

ANALYSIS : The chart indicates that total sales vary significantly across outlet location types, with Tier 3 outlets generating the highest sales, followed by Tier 2, while Tier 1 records the lowest. This pattern highlights that outlets in higher-tier locations contribute more strongly to overall revenue.The insights suggest that Tier 3 outlets represent the most profitable markets, making them crucial for maintaining and expanding sales strategies.
